## 0. Imports

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

## 1. Load Pre-Processed Data

```python
file_path = '../data/cleaned_data.xlsx'
clean_df = pd.read_excel(file_path, header=0)
print(f'Dataset shape: {clean_df.shape}')
print(f'Columns: {list(clean_df.columns)}')
clean_df.head()
```

## 2. Descriptive Statistics

Before plotting, we examine the summary statistics split by default status. This gives an at-a-glance comparison of how defaulters differ from non-defaulters across key financial variables.

```python
# Side-by-side descriptive stats for defaulters vs non-defaulters
key_cols = ['LIMIT_BAL', 'AGE', 'BILL_AMT1', 'PAY_AMT1']
print('=== Non-Defaulters (default=0) ===')
print(clean_df[clean_df['default'] == 0][key_cols].describe().round(2))
print()
print('=== Defaulters (default=1) ===')
print(clean_df[clean_df['default'] == 1][key_cols].describe().round(2))
```

## 3. Label Encoding for Readable Plots

```python
education_map = {1: 'Graduate School', 2: 'University', 3: 'High School', 4: 'Others'}
marriage_map  = {1: 'Married', 2: 'Single', 3: 'Others'}
gender_map    = {1: 'Male', 2: 'Female'}
default_map   = {1: 'Default', 0: 'No Default'}

clean_df['EDUCATION_label'] = clean_df['EDUCATION'].map(education_map)
clean_df['MARRIAGE_label']  = clean_df['MARRIAGE'].map(marriage_map)
clean_df['SEX_label']       = clean_df['SEX'].map(gender_map)
clean_df['default_label']   = clean_df['default'].map(default_map)
```

## 4. Feature Engineering

We derive four new features that summarise a customer's financial behaviour over the 6-month window. These are more informative than the raw monthly columns individually.

```python
bill_cols = ['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6']
pay_cols  = ['PAY_AMT1',  'PAY_AMT2',  'PAY_AMT3',  'PAY_AMT4',  'PAY_AMT5',  'PAY_AMT6']

# Average bill and payment over 6 months
clean_df['AVG_BILL']    = clean_df[bill_cols].mean(axis=1)
clean_df['AVG_PAYMENT'] = clean_df[pay_cols].mean(axis=1)

# Credit utilisation: how much of the limit is being used on average
clean_df['UTILIZATION'] = clean_df['AVG_BILL'] / clean_df['LIMIT_BAL']

# Payment-to-bill ratio: how much of what is owed is actually being paid
# Adding 1e-9 avoids division by zero for customers with no bill
clean_df['PAY_TO_BILL_RATIO'] = clean_df['AVG_PAYMENT'] / (clean_df['AVG_BILL'] + 1e-9)

# Bill trend: BILL_AMT1 is the most recent month, BILL_AMT6 is the oldest
# A ratio > 1 means debt is growing; < 1 means debt is shrinking
clean_df['BILL_TREND'] = clean_df['BILL_AMT1'] / (clean_df['BILL_AMT6'] + 1e-9)

print('Derived features added:')
print(clean_df[['LIMIT_BAL', 'AVG_BILL', 'UTILIZATION', 'PAY_TO_BILL_RATIO', 'BILL_TREND', 'default']].head())
```

---
## Plot 1 — Class Distribution (Target Variable)

**Observation:** The dataset is imbalanced, with approximately 78% of clients not defaulting and 22% defaulting. This ~3.5:1 ratio is moderate but meaningful — a naive classifier that always predicts "No Default" would already achieve 78% accuracy, making raw accuracy a misleading metric. 

**Modeling Implication:** F1-score and ROC-AUC should be the primary evaluation metrics. Techniques such as class weighting or SMOTE may be required to prevent models from being biased toward the majority class.

```python
class_counts = clean_df['default_label'].value_counts().reset_index()
class_counts.columns = ['Status', 'Count']
class_counts['Percentage'] = (class_counts['Count'] / class_counts['Count'].sum() * 100).round(1)
class_counts['Label'] = class_counts['Percentage'].astype(str) + '%'

fig = px.bar(
    class_counts,
    x='Status',
    y='Count',
    color='Status',
    text='Label',
    title='Plot 1: Class Distribution — Default vs No Default',
    color_discrete_map={'No Default': '#2ecc71', 'Default': '#e74c3c'},
    labels={'Status': 'Default Status', 'Count': 'Number of Clients'}
)
fig.update_traces(textposition='outside', textfont_size=14)
fig.update_layout(showlegend=False, yaxis_range=[0, class_counts['Count'].max() * 1.15])
fig.show()
```

---
## Plot 2 — Demographic Overview: Default Rate Across SEX, EDUCATION & MARRIAGE

**Observation:** Default rates are remarkably consistent across all demographic subgroups, hovering between 19% and 25% regardless of gender, education level, or marital status. No single demographic category shows a dramatically elevated or reduced default risk.

**Modeling Implication:** These categorical features are unlikely to be strong standalone predictors and will probably rank low in feature importance for tree-based models. However, they may contribute through interaction effects with financial variables such as `LIMIT_BAL` and `AGE`, so they should be retained in the feature set rather than dropped.

```python
def get_default_rate(df, group_col, label_col):
    """Calculate default rate (%) per group."""
    grp = df.groupby(label_col)['default'].mean().reset_index()
    grp.columns = ['Group', 'Default Rate']
    grp['Default Rate (%)'] = (grp['Default Rate'] * 100).round(1)
    grp['Category'] = group_col
    return grp

sex_rates  = get_default_rate(clean_df, 'Sex',       'SEX_label')
edu_rates  = get_default_rate(clean_df, 'Education', 'EDUCATION_label')
mar_rates  = get_default_rate(clean_df, 'Marriage',  'MARRIAGE_label')

fig = make_subplots(
    rows=1, cols=3,
    subplot_titles=['By Sex', 'By Education', 'By Marriage Status']
)

for col_idx, (df_grp, color) in enumerate(
    zip([sex_rates, edu_rates, mar_rates], ['#3498db', '#9b59b6', '#e67e22']), start=1
):
    fig.add_trace(
        go.Bar(
            x=df_grp['Group'],
            y=df_grp['Default Rate (%)'],
            text=df_grp['Default Rate (%)'].astype(str) + '%',
            textposition='outside',
            marker_color=color,
            showlegend=False
        ),
        row=1, col=col_idx
    )

fig.update_layout(
    title_text='Plot 2: Default Rate (%) Across Demographic Groups',
    height=420,
    yaxis=dict(range=[0, 35], title='Default Rate (%)'),
    yaxis2=dict(range=[0, 35]),
    yaxis3=dict(range=[0, 35])
)
fig.show()
```

---
## Plot 3 — The Risk Staircase: Default Rate by Most Recent Payment Status (PAY_1)

**Payment Status Key:** -2 = no consumption; -1 = paid in full; 0 = revolving balance paid; 1+ = months overdue.

**Observation:** This is the single strongest signal in the dataset. Clients in good standing (-2, -1, 0) default at a low baseline rate of 12–17%. However, default risk increases sharply and monotonically with each month of delay — jumping from ~34% at one month late to over 75% at three months late.

**Modeling Implication:** `PAY_1` (and its lagged equivalents `PAY_2`–`PAY_6`) will almost certainly be the top-ranked features in any tree-based model. This monotonic step-up pattern also makes logistic regression and SVM viable, as the relationship is highly structured and directional.

```python
risk_df = clean_df.groupby('PAY_1')['default'].mean().reset_index()
risk_df['Default Risk (%)'] = (risk_df['default'] * 100).round(1)
# Filter to the meaningful range (-2 to 4 months late)
risk_df = risk_df[risk_df['PAY_1'] <= 4]

fig = px.bar(
    risk_df,
    x='PAY_1',
    y='Default Risk (%)',
    title='Plot 3: The Risk Staircase — Default Rate by Most Recent Payment Status',
    labels={'PAY_1': 'Payment Status (≤0: On Time / Paid, 1+: Months Overdue)'},
    text_auto='.1f',
    color='Default Risk (%)',
    color_continuous_scale='Reds'
)
fig.update_traces(textposition='outside')
fig.update_layout(yaxis_range=[0, 90], coloraxis_showscale=False)
fig.show()
```

---
## Plot 4 — Credit Limit vs Default Status

**Observation:** Non-defaulters tend to have noticeably higher credit limits than defaulters. The median credit limit for non-defaulters is substantially above that of defaulters, and defaulters show a tighter, lower-range distribution. This suggests that the bank had already assigned lower limits to higher-risk customers, making `LIMIT_BAL` a useful proxy for pre-assessed creditworthiness.

**Modeling Implication:** `LIMIT_BAL` is a strong continuous predictor. Its relationship with the target is not perfectly linear — there is overlap between the groups — so non-linear models (e.g., Decision Tree, SVM with RBF kernel) may capture this boundary more effectively than Logistic Regression.

```python
fig = px.box(
    clean_df,
    x='default_label',
    y='LIMIT_BAL',
    color='default_label',
    title='Plot 4: Credit Limit Distribution by Default Status',
    labels={'default_label': 'Default Status', 'LIMIT_BAL': 'Credit Limit (NT$)'},
    color_discrete_map={'No Default': '#2ecc71', 'Default': '#e74c3c'},
    points='outliers'
)
fig.update_layout(showlegend=False)
fig.show()
```

---
## Plot 5 — Correlation Heatmap (Top Features vs Default)

**Observation:** The `PAY_X` repayment status columns show the strongest positive correlations with `default` (r ≈ 0.3–0.4), confirming the Risk Staircase finding. `LIMIT_BAL` shows a moderate negative correlation (r ≈ -0.15), meaning higher limits are associated with lower default risk. The `BILL_AMT` columns are highly intercorrelated with each other (r > 0.9), indicating multicollinearity — using all six adds little extra information.

**Modeling Implication:** For models sensitive to multicollinearity (Logistic Regression, kNN), it is advisable to use only `BILL_AMT1` or the engineered `AVG_BILL` rather than all six raw bill columns. Tree-based models are unaffected by this redundancy.

```python
# Select a focused set of features — drop label columns and low-signal demographics
focus_cols = [
    'LIMIT_BAL', 'AGE',
    'PAY_1', 'PAY_2', 'PAY_3',
    'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3',
    'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3',
    'UTILIZATION', 'PAY_TO_BILL_RATIO', 'BILL_TREND',
    'default'
]

corr_matrix = clean_df[focus_cols].corr(numeric_only=True)

fig = px.imshow(
    corr_matrix,
    text_auto='.2f',
    aspect='auto',
    color_continuous_scale='RdBu_r',
    title='Plot 5: Correlation Heatmap — Key Financial Features & Engineered Variables'
)
fig.update_layout(
    width=850,
    height=750,
    xaxis_tickangle=-45
)
fig.show()
```

---
## Bonus Plot — Average Bill Trend Over 6 Months

**Observation:** Both defaulters and non-defaulters show broadly similar bill amounts over the 6-month window, but defaulters consistently carry slightly higher average bills throughout. Notably, both groups show a dip in BILL_AMT3–AMT4 before rising again toward the most recent month (AMT1), suggesting a seasonal or behavioural pattern in spending.

**Modeling Implication:** The raw monthly bill columns carry some signal individually, but the derived `AVG_BILL` and `BILL_TREND` features likely capture this trajectory more compactly. The overlap between groups also confirms that bill amount alone is not sufficient — payment behaviour (PAY_X) must be combined with it for strong prediction.

```python
# Order from oldest (BILL_AMT6) to most recent (BILL_AMT1)
bill_cols_ordered = ['BILL_AMT6', 'BILL_AMT5', 'BILL_AMT4', 'BILL_AMT3', 'BILL_AMT2', 'BILL_AMT1']
month_labels      = ['Month -6', 'Month -5', 'Month -4', 'Month -3', 'Month -2', 'Month -1 (Latest)']

avg_bills = clean_df.groupby('default_label')[bill_cols_ordered].mean().reset_index()
melted    = avg_bills.melt(id_vars='default_label', var_name='Month_Raw', value_name='Average Bill (NT$)')

# Map raw column names to readable month labels
label_map = dict(zip(bill_cols_ordered, month_labels))
melted['Month'] = melted['Month_Raw'].map(label_map)
melted['Month'] = pd.Categorical(melted['Month'], categories=month_labels, ordered=True)
melted = melted.sort_values('Month')

fig = px.line(
    melted,
    x='Month',
    y='Average Bill (NT$)',
    color='default_label',
    markers=True,
    title='Bonus: Average Bill Amount Trend Over 6 Months (Oldest → Most Recent)',
    color_discrete_map={'No Default': '#2ecc71', 'Default': '#e74c3c'},
    labels={'default_label': 'Default Status'}
)
fig.update_layout(xaxis_title='Time Period', legend_title='Default Status')
fig.show()
```

## EDA Summary & Modeling Implications

| Finding | Key Feature(s) | Modeling Implication |
|---|---|---|
| Dataset is imbalanced (78/22) | `default` | Use F1 & ROC-AUC; consider class weighting |
| Demographic features show weak signal | `SEX`, `EDUCATION`, `MARRIAGE` | Retain for interactions; don't rely on them alone |
| Payment delay is the strongest risk signal | `PAY_1`–`PAY_6` | Will rank as top features in tree models |
| Higher credit limit = lower default risk | `LIMIT_BAL` | Strong continuous predictor for all models |
| BILL_AMT columns are highly intercorrelated | `BILL_AMT1`–`BILL_AMT6` | Use `AVG_BILL` to reduce multicollinearity |
| Engineered features capture financial behaviour | `UTILIZATION`, `PAY_TO_BILL_RATIO`, `BILL_TREND` | Include in all model feature sets |

**Recommended classifiers based on EDA:**
- **Logistic Regression** — good baseline; will benefit from scaled features and reduced multicollinearity
- **Decision Tree / Random Forest** — well-suited to the step-wise `PAY_1` signal and feature interactions
- **SVM or kNN** — viable with scaled features; SVM with RBF kernel may capture non-linear `LIMIT_BAL` boundary well
